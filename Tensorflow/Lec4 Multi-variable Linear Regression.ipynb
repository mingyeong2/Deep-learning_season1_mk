# Hypothesis not using matrix

```python
import tensorflow as tf
```

```python
#data
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
```

```python
#placeholders for a tensor that will be always fed

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)
```

```python
w1 = tf.Variable(tf.random_normal([1]), name = 'weight1')
w2 = tf.Variable(tf.random_normal([1]), name = 'weight2')
w3 = tf.Variable(tf.random_normal([1]), name = 'weight3')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
hypothesis = w1 * x1 + w2*x2 + w3*x3 + b
```

```python
#cost/loss function 정의
cost = tf.reduce_mean(tf.square(hypothesis - y_data))
```

```python
#Minimize! 
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
#graph를 session에 launch
sess = tf.Session()
```

```python
#Initialize global variables in the graph
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {x1:x1_data, x2:x2_data, x3:x3_data, 
                                              Y : y_data})

    if step % 10 == 0:
        print('Step: ', step, 'Cost: ', cost_val, '\nPrediction:\n', hy_val)
```

# Hypothesis using Matrix

```python
#data를 matrix로!
x_data = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.],
          [96., 98., 100.],[73., 66., 70.]]
y_data = [[152.],[185.],[180.],[196.],[142.]]
```

```python
#placeholder for a tensor
#None을 통해서 instance가 늘어나도 shape을 유지하도로 함. (numpy의 -1)
X = tf.placeholder(tf.float32, shape = [None, 3])
Y = tf.placeholder(tf.float32, shape = [None, 1])

W = tf.Variable(tf.random_normal([3,1]), name = 'weight')
b = tf.Variable(tf.random_normal([1]), name = 'bias')
```

```python
#hypothesis: XW + b
hypothesis = tf.matmul(X,W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost,hypothesis, train],
                                  feed_dict = {X:x_data, Y:y_data})
    if step % 10 == 0:
        print('step: ', step, ', cost: ', cost_val, 
             '\nPrediction: \n', hy_val)
```

# Loading Data from File

```python
import numpy as np
```

```python
#data를 받아서 x와 y값으로 분리
xy = np.loadtxt('data-01-test-score.csv', delimiter = ',', dtype = np.float32)
x_data = xy[:,0:-1]
y_data = xy[:,[-1]]
```

```python
#data 확인
print(x_data.shape, x_data, len(x_data))
print(y_data.shape, y_data, len(y_data))
```

Gradient Descent

```python
#placeholder... set the shape as same as the original
x_attr = x_data.shape[1]
y_shape = y_data.shape[1]

X = tf.placeholder(tf.float32, shape = [None, x_attr])
Y = tf.placeholder(tf.float32, shape = [None, y_shape])

W = tf.Variable(tf.random_normal([x_attr,y_shape]), name = 'weight')
b = tf.Variable(tf.random_normal([y_shape]), name = 'bias')
```

```python
#hypothesis
hypothesis = tf.matmul(X,W) + b
```

```python
#Simplified cost/loss function
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train], feed_dict = {X :x_data, Y : y_data})
    if step % 40 == 0:
        print('step: ', step, ' | cost: ', cost_val, '\nPrediction: \n', np.concatenate(hy_val))
        
print('\nY data: ', np.concatenate(y_data))
```

## 파일이 너무 큰 경우: Queue Runners
여러개의 파일을 불러 올 경우, Filename Queue에 filename을 쌓고,
이를 각각에 맞게 encoding한 후 다시 Queue를 하면 tensor가 알아서 파일을 읽어준다

```python
# 1. 가지고 있는 파일들의 list를 만들어 주기
filename_queue = tf.train.string_input_producer(['data-01-test-score.csv'], 
                                                shuffle = False, name = 'filename_queue')
```

```python
# 2. file을 읽어올 reader를 정의하기
reader = tf.TextLineReader()
key, value = reader.read(filename_queue)
```

```python
# 3.읽은 value를 parsing 종류에 따라서 읽어주기
record_defaults = [[0.],[0.],[0.],[0.]]  #값이 없을 경우
xy = tf.decode_csv(value, record_defaults = record_defaults)
```

이 값들을 batch를 통해서 불러오기

```python
#collect batches of csv in 
train_x_batch, train_y_batch = tf.train.batch([xy[0:-1], xy[-1:]], batch_size =10)  #한 번에 10개씩 가져와라!
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
#Minimize
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 1e-5)
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
#Start populating the filename queue (general)
coord = tf.train.Coordinator()
threads = tf.train.start_queue_runners(sess = sess, coord = coord)
```

```python
for step in range(2001):
    x_batch, y_batch = sess.run([train_x_batch, train_y_batch]) #이를 feed_dict를 통해 model에 넣어주기
    
    cost_val, hy_val, _ = sess.run([cost, hypothesis, train],
                                  feed_dict = {X:x_batch, Y:y_batch})
    if step % 40 == 0:
        print('Batch:  ' , x_batch, y_batch)
        print('step: ', step, ' | cost: ', cost_val, '\nPrediction: \n', np.concatenate(hy_val))
        
print('\nY data: ', np.concatenate(y_data))
    

#general
coord.request_stop()
coord.join(threads)
```
