```python
import torch
from torch import nn, optim
import torch.utils.data as Data
from torch.autograd import Variable
```

```python
class LinearRegression(nn.Module):
    
    def __init__(self, input_dim, output_dim, bias = True):
        super(LinearRegression, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim, bias)
        
    def forward(self, x):
        out = self.linear(x)
        return out
        
```

```python
x_data = [[1,2],[2,3],[3,1],[4,3],[5,3],[6,2]]
y_data = [[0],[0],[0],[1],[1],[1]]

input_dim = len(x_data[0])
output_dim = len(y_data[0])
```

```python
X = Variable(torch.tensor(x_data)).type(torch.float32)
Y = Variable(torch.tensor(y_data)).type(torch.float32)
```

```python
linear = LinearRegression(input_dim, output_dim)


#Sigmoid
sigmoid = nn.Sigmoid()

lr = 0.01
optimiser = torch.optim.SGD(linear.parameters(), lr = lr)
```

```python
for step in range(10001):
    
    optimiser.zero_grad()
    
    outputs = linear.forward(X)
    outputs = sigmoid(outputs)
    cost = -(Y * torch.log(outputs) + (1 - Y)* torch.log(1 - outputs)).mean()
    cost.backward(retain_graph=True)
    optimiser.step()
    
    if step % 100 == 0:
        print('parameters: ', end = '')
        for i in linear.parameters():
            print(i.data.numpy())
        print('step {}, loss {}, \nPrediction:{}'.format(step, cost.item(), outputs.data.numpy()))
        print( )
        
predicted = (linear.forward(X) > 0.5).float()
accuracy = (predicted == Y).float().mean()

print('Prediction: ', predicted, '\n Accuracy: ', accuracy.data)
```

## Classifying diabetes

```python
import numpy as np
```

```python
xy = np.loadtxt('data-03-diabetes.csv', delimiter = ',', dtype = np.float32)
x_data = xy[:,:-1]
y_data = xy[:,[-1]]

input_dim = len(x_data[0])
output_dim = len(y_data[0])

X = Variable(torch.tensor(x_data)).float()
Y = Variable(torch.tensor(y_data)).float()
```

```python
linear = nn.Linear(input_dim, output_dim, bias = True)
sigmoid = nn.Sigmoid()
model = nn.Sequential(linear, sigmoid)

lr = 0.01
optimiser = optim.SGD(model.parameters(), lr = lr)
```

```python
for step in range(10001):
    optimiser.zero_grad()
    
    outputs = model(X)
    cost = -(Y*torch.log(outputs) + (1-Y)*torch.log(1-outputs)).mean()
    cost.backward(retain_graph = True)
    optimiser.step()
    
    if step % 100 == 0:
        print('parameters: ', end = '')
        for i in linear.parameters():
            print(i.data.numpy())
        print('step {}, loss {}, \nPrediction:{}'.format(step, cost.item(), outputs.data.numpy()))
        print( )
        
predicted = (linear.forward(X) > 0.5).float()
accuracy = (predicted == Y).float().mean()

print('Prediction: ', predicted, '\n Accuracy: ', accuracy.data)
```
