```python
import torch
from torch import nn, optim
import torch.nn.functional as F
from torch.autograd import Variable
import numpy as np
```

```python
x_train = [[73., 80., 75.], [93., 88., 93.], [89., 91., 90.],[96., 98., 100.],[73.,66.,70.]]
y_train = [[152.],[185.],[180.],[196.],[142.]]

trainX = Variable(torch.tensor(x_train))
trainY = Variable(torch.tensor(y_train))
```

```python
class MultiVariableLinearRegression(nn.Module):
    
    def __init__(self, input_dim, output_dim, bias = True):
        super(MultiVariableLinearRegression, self).__init__()
        self.linear = nn.Linear(input_dim, output_dim, bias)
        
    def forward(self, x):
        out = self.linear(x)
        return out
```

```python
#hypothesis 
model = MultiVariableLinearRegression(3, 1, bias = True)

#cost: MSE
criterion = nn.MSELoss()

learning_rate = 1e-5
#minimize
optimiser = torch.optim.SGD(model.parameters(), lr = learning_rate)
```

### Traing the Model

```python
epochs = 2000

for epoch in range(1, epochs + 1):
    
    #Clears the gradients of all optimized  tensors (set gradients as 0)
    optimiser.zero_grad()
    
    #forward to get predicted value
    outputs = model.forward(trainX)
    loss = criterion(outputs, trainY)
    loss.backward() #apply backpropagation
    optimiser.step()  #update the parameters
    if epoch % 20 == 0:
        print('parameters: ', end = '')
        for i in model.parameters():
            print(i.data.numpy())
        print('epoch {}, loss {}, \nPrediction:{}'.format(epoch, loss.item(), outputs.data.numpy()))
        print( )
```

## Loading Data from file

```python
xy = np.loadtxt('data-01-test-score.csv', delimiter = ',', dtype = np.float32)
```

```python
x = xy[:,:-1]
y = xy[:,[-1]]  #마지막 값만!
```

```python
train_X = Variable(torch.tensor(x, dtype = torch.float32))
train_Y = Variable(torch.tensor(y, dtype = torch.float32))
```

```python
model = MultiVariableLinearRegression(len(x[0]), len(y[0]), bias = True)

criterion = nn.MSELoss()
lr = 1e-5
optimiser = optim.SGD(model.parameters(), lr = lr)
```

```python
epochs = 2000

for epoch in range(1, epochs):
    
    optimiser.zero_grad()
    
    outputs = model.forward(train_X)
    loss = criterion(outputs, train_Y)
    loss.backward()
    optimiser.step()
    
    if epoch % 200 == 0:
        for i in model.parameters():
            print(i.data.numpy())
        print('epoch  {},  loss {},  \nPrediction: {}'.format(epoch, loss, outputs.data.numpy()))
        
print('\nreal Y: \n', y)
```
